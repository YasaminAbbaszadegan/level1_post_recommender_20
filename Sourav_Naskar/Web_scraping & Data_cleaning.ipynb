```python
import requests
import bs4 as bs
from selenium import webdriver
import csv
from selenium.webdriver.chrome.options import Options
import time
import re
import string
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords       # add "nltk.download('stopwords')" if not downloaded
from nltk.stem import PorterStemmer
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer # add "nltk.download('wordnet')" if not downloaded
import pandas as pd
```

```python
url = 'https://discuss.codecademy.com/c/community/1406'
```

```python
PATH="C:\Program Files (x86)\chromedriver.exe"
Chrome_Options=webdriver.ChromeOptions()
Chrome_Options.add_argument("--incognito")         #runs in incognito mode
Chrome_Options.add_argument("--headless")          #runs in background
driver=webdriver.Chrome(PATH,options=Chrome_Options)
driver.implicitly_wait(2)
driver.get(url)
```

```python
for i in range(50): #scrolls down 50 times 
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(0.5)
    print(i)
```

```python
# Scraping the whole page 
page = driver.execute_script('return document.body.innerHTML')
soup = bs.BeautifulSoup(''.join(page), 'html.parser')
```

```python
posts = soup.find_all('a', class_= 'title raw-link raw-topic-link')
links = []
titles = []
for post in posts:
    titles.append(post.get_text())
    links.append(post.get('href'))
length=len(titles)
print(titles)
print(length)
```

```python
posts = soup.find_all('span', class_= 'category-name')
categories=[]

for post in posts:
    categories.append(post.get_text())
length=len(categories)
categories.pop(0)
print(length)
print(categories)
```

```python
posts = soup.find_all('span', class_= 'number')
replies=[]
for post in posts:
    replies.append(post.get_text())
length=len(replies)

print(length)
print(replies)
```

```python
views=[]
for i in range(1,len(replies),2):    #to make the length of replies half
    views.append(replies[i])
    replies[i]=None
replies = [i for i in replies if i] 
print(len(views))
print(len(replies))
```

```python
list_of_comments = []
for elem in links:
    try:
        driver.get("https://discuss.codecademy.com"+elem)
        time.sleep(1) #loading 
        page = driver.execute_script('return document.body.innerHTML')
        soup = bs.BeautifulSoup(''.join(page),'html.parser')
        comments = soup.find_all('div', class_= 'cooked')
        temp = ''
        for ele in comments:
            temp += ele.get_text()
        list_of_comments.append(temp)
    except:
        pass
```

```python
for i in list_of_comments:
    x=i.split('\n')
    print(x)
    print(len(x))
```

```python
def clean_data(data):
    cleaned_data = [] 
    for elem in data:
        temp = re.sub('[\(\)\{\}<>\-“”]', '', elem) 
        temp = re.split('\. |! |\n |\? ', temp.strip())
        cleaned_data.extend(temp)



    #actual cleaning data
    for i in range(len(cleaned_data)):

        #clear puncatuation, tokenizing the sentence and make lower case
        cleaned_data[i] =  cleaned_data[i].translate(str.maketrans('', '', string.punctuation)) #clear punctuation
        cleaned_data[i] = cleaned_data[i].lower().split()


        #delete stopwords, and lemmatize the words
        stop_words = set(stopwords.words('english'))
        ps = PorterStemmer()
        lemmatizer = WordNetLemmatizer()
        temp = []
        for elem in cleaned_data[i]:
            if elem not in stop_words and not elem.startswith('http'):
                #elem = ps.stem(elem) this is stemming
                temp.append(lemmatizer.lemmatize(elem))
        cleaned_data[i] = temp

    
    res = [ele for ele in cleaned_data if ele != []] 
    #this loop is to prevent jupyter from returning error for outputing too fast
    i = 0
    while i < len(res):
        for j in range(1000):
            if i == len(res):
                break
            print(res[i])
            i=i+1
        time.sleep(1)
```

```python
clean_data(titles)
```

```python
clean_data(list_of_comments)
```

```python
data_df = pd.DataFrame({'Title': titles,
'Category': categories,
'Replies': replies,
'Views': views,'Comments':list_of_comments})
print(data_df.info())
data_df.head(10)
```

```python
data_df.to_csv("data.csv")
```

```python
# Let's take a look at the updated data
data_clean = pd.DataFrame(data_df)
data_clean
```

```python
# We are going to create a document-term matrix using CountVectorizer, and exclude common English stop words
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(data_clean.Comments)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```
